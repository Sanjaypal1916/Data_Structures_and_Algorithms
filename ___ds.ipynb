```python
class Node:
    def __init__(self, data, next):
        self.data = data
        self.next = next

class linkedlist:
    def __init__(self):
        self.head = None

    def printLL(self):
        if self.head is None:
            raise Exception("LinkedList is empty") 
            return
        itr = self.head
        lltr = ""
        while itr:
            lltr += str(itr.data) + "--->" if itr.next else str(itr.data)
            itr = itr.next
        print(lltr)

    def insert_at_begining(self, data):
        node = Node(data, self.head)
        self.head = node
    
    def insert_at_end(self, data):
        node = Node(data, None)
        if self.head is None:
            self.head = node
        itr = self.head
        while itr.next:
            itr = itr.next 
        itr.next = node
        
    
    def length(self):
        itr = self.head
        count = 0 
        while itr:
            count += 1
            itr = itr.next
        print(count)
        return count

    def insert_at(self, index, data ):

        if index < 0  or index > self.length():
             print("invalid")
             return 
         
        if index == 0 :
             self.insert_at_begining(data)
             return
        count=0
        itr = self.head
        while itr and count != index-1:
            count += 1
            itr = itr.next
        node = Node(data, itr.next)
        itr.next = node

    def remove_at(self, index):
        if index < 0  or index > self.length():
            print("invalid")
            return
        
        if index == 0 :
            self.head = self.head.next
            return

        count = 0 
        itr = self.head
        while itr and count != index-1:
            count += 1
            itr = itr.next
        itr.next = itr.next.next
        return
    
    def insert_list(self, list):
        for i in list:
            self.insert_at_end(i)


    def insert_after_value(self, data_after, data_to_insert):
        itr = self.head
        while itr:
            data = itr.data
            if data == data_after:
                break
            itr = itr.next
        node = Node(data_to_insert, itr.next)
        itr.next = node

    def remove_by_value(self, data_to):
        # Remove first node that contains data
        itr = self.head
        while itr.next:
            data = itr.next.data
            if data == data_to:
                break
            itr = itr.next
        itr.next = itr.next.next

    
        
```

```python
class Node:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev
    
class LinkedList:
    def __init__(self):
        self.head = None
    
    def insert_at_begining(self, data):
        if self.head is None:  # case: empty list
            self.head = Node(data, None, None)
        else:
            node = Node(data, self.head, None)
            self.head.prev = node
            self.head = node
        

    def insert_at_end(self, data):
        itr = self.head
        while itr.next:
            itr = itr.next
        node = Node(data, None, itr)
        itr.next = node

    def printLL(self):
        if self.head is None:
            raise Exception("LinkedList is empty") 
            return
        itr = self.head
        lltr = ""
        while itr:
            lltr += str(itr.data) + "--->" if itr.next else str(itr.data)
            itr = itr.next
        print(lltr)

    def print_forward(self):
        itr = self.head
        lltr = ""
        while itr:
            lltr += str(itr.data) + "<--->" if itr.next else str(itr.data) 
            itr = itr.next 
        print(lltr)

    def print_backward(self):
        itr = self.head
        lltr = ""
        while itr.next:
            itr = itr.next

        while itr:
            lltr += str(itr.data) + "<--->" if itr.prev else str(itr.data) 
            itr = itr.prev
        print(lltr)

    
```

```python
link = LinkedList()
link.insert_at_begining(1)
link.insert_at_begining(2)
link.insert_at_begining(3)
link.insert_at_begining(4)
link.printLL()

link.insert_at_end(9)
link.insert_at_end(10)
link.insert_at_end(11)
link.printLL()

link.print_backward()
link.print_forward()
```

```python
class HashTable:
    def __init__(self):
        self.MAX = 10
        self.arr = [None for i in range(10)]

    def get_hash(self, key):
        hash = 0 
        for k in key:
            hash += ord(k)
        return hash % self.MAX
    
    def __getitem__(self, key):
        h = self.get_hash(key)
        return self.arr[h]

    def __setitem__(self, key, val):
        h = self.get_hash(key)
        self.arr[h] = val
        return True
```

## Solving the Problem of Hash Collision


### Linear Chaining  --> O(n)

```python
class HashTable:
    def __init__(self):
        self.MAX = 10
        self.arr = [[] for i in range(10)]

    def get_hash(self, key):
        hash = 0 
        for k in key:
            hash += ord(k)
        return hash % self.MAX
    
    def __getitem__(self, key):
        h = self.get_hash(key)
        for k, v in enumerate(self.arr[h]):
            if k == key:
                return v
        return "not found"

    # def __setitem__(self, key, val):
    #     h = self.get_hash(key)
    #     self.arr[h].append((key, val))            
    #     return True
    
    # set and update logic
    def __setitem__(self, key, val):
        h = self.get_hash(key)
        found = False
        for index, tup in enumerate(self.arr[h]):
            if len(tup) == 2 and tup[0] == key:
                self.arr[h][index] = (key, val)
                found = True
                break
        if not found:
            self.arr[h].append((key, val))

    def __delitem__(self, key):
        h = self.get_hash(key)
        for idx, kv in enumerate(self.arr[h]):
            if kv[0] == key and len(kv) == 2:
                print(f"deleted : {kv[0]} , {kv[1]}")
                del self.arr[h][idx]

        
```

### Linear Probing 

```python
class Hashtable:
    def __init__(self):
        self.MAX = 10
        self.arr = [None for i in range(10)]

    def get_hash(self, key):
        h = 0
        for char in key:
            h += ord(char)
        return h % 10

    def __setitem__(self, key, val):
        h = self.get_hash(key)
        start = h 
        while self.arr[h]:
            if self.arr[h][0] == key:
                self.arr[h] = (key, val)
            h = (h+1) % 10
            if h == start:
                print("hashtable full")
                break 
        self.arr[h] = (key,val)
        

    def __getitem__(self, key ):
        h = self.get_hash(key)
        start = h 
        while self.arr[h]:
            if self.arr[h][0] == key:
                return self.arr[h][1]
            h = (h+1) % 10
            if h == start:
                print("does Not exist ")
                break
```

```python
ht = Hashtable()
ht["apple"] = 100
ht["banana"] = 200
ht["apple"] = 300  # update
print(ht["apple"])  # 300
print(ht["banana"]) 
```
